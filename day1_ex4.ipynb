```python
class Pyramidal:
  def __init__(self, gid):
    self._gid = gid
    self._load_morphology()
  def __repr__(self):
    return f'Pyramidal[{self._gid}]'
  def _load_morphology(self):
    cell = h.Import3d_SWC_read()
    cell.input('c91662.CNG.swc')
    i3d = h.Import3d_GUI(cell, False)
    i3d.instantiate(self)
```

```python
pyr1 = Pyramidal(0)
```
